# Machine Learning
- List of features
- List of predicted values (normally one item)
- Train/Test split
- Import and instantiate linear model
- Apply fit model with the Train values

```python
from sklearn.model_selection import train_test_split
# not used: 'longitude', 'latitude', 'ocean_proximity'

X = housing_data[['housing_median_age', 'total_rooms',
    'total_bedrooms', 'population', 'households', 'median_income']]
y = housing_data['median_house_value']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X=X_train, y=y_train)
```

```python
print(lm.intercept_)
```

```python
coeff = pd.DataFrame(lm.coef_, X.columns, columns=["Coeff"])
coeff
```

```python
prediction = lm.predict(X_test)
```

```python
sns.scatterplot(x=y_test, y=prediction)
```
